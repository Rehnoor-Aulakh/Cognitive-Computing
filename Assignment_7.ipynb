```python
# Q1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(102317156)
```

```python
# Part I
sales_data = np.random.randint(1000, 5001, size=(12, 4))
columns = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df = pd.DataFrame(sales_data, columns=columns, index=months)
print(df)
```

```python
# Part II
print("First 5 rows:\n", df.head())
print("\nSummary Statistics:\n", df.describe())
```

```python
totalSalesCategory = df.sum()
print("\nTotal Sales per Category:\n", totalSalesCategory)
```

```python
df["Total Sales"] = df.sum(axis=1)
df["Growth Rate"] = df["Total Sales"].pct_change() * 100

roll_number = 102317156
if roll_number % 2 == 0:
    df["Electronics"] *= 0.9
else:
    df["Clothing"] *= 0.85

print(df)
```

```python
# Part III:
plt.figure(figsize=(10, 6))
sns.lineplot(data=df[columns])
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend(columns)
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df[columns])
plt.title("Sales Distribution by Category")
plt.show()
```

```python
# Q2
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])
flat = array.flatten()
abs_values = np.abs(array)
percentiles_flattened = np.percentile(flat, [25, 50, 75])
percentiles_columns = np.percentile(array, [25, 50, 75], axis=0)
percentiles_rows = np.percentile(array, [25, 50, 75], axis=1)

mean_flattened = np.mean(flat)
mean_columns = np.mean(array, axis=0)
mean_rows = np.mean(array, axis=1)

median_flattened = np.median(flat)
median_columns = np.median(array, axis=0)
median_rows = np.median(array, axis=1)

std_flattened = np.std(flat)
std_columns = np.std(array, axis=0)
std_rows = np.std(array, axis=1)

print("Absolute Values:\n", abs_values)
print("Percentiles (Flattened):\n", percentiles_flattened)
print("Percentiles (Columns):\n", percentiles_columns)
print("Percentiles (Rows):\n", percentiles_rows)

print("Mean (Flattened):\n", mean_flattened)
print("Mean (Columns):\n", mean_columns)
print("Mean (Rows):\n", mean_rows)

print("Median (Flattened):\n", median_flattened)
print("Median (Columns):\n", median_columns)
print("Median (Rows):\n", median_rows)

print("Standard Deviation (Flattened):\n", std_flattened)
print("Standard Deviation (Columns):\n", std_columns)
print("Standard Deviation (Rows):\n", std_rows)
```
