```python
import pandas as pd
import numpy as np

first_name_letter='R'
last_name_letter='A'
X=ord(first_name_letter)+ord(last_name_letter)
sales=np.array([X,X+50,X+100,X+150,X+200])
tax_rate=((sales%5)+5)/100
sales_after_tax=sales-(sales*tax_rate)
discounted_sales=np.where(sales<X+100,sales*0.95,sales*0.90)
discount=sales
df={
    "sales":sales,
    "sales_after_tax":sales_after_tax,
    "discounted_sales":discounted_sales,
}

sales_weeks = np.tile(sales, (3, 1))  # Create 3x5 matrix

# Increase sales by 2% per week
week_multipliers = np.array([1.00, 1.02, 1.04]).reshape(3, 1)
adjusted_sales_weeks = sales_weeks * week_multipliers
print("Sales for multiple weeks:", adjusted_sales_weeks)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-10,10,100)
print(x)
y_square=x**2
y_sin=np.sin(x)
y_exp=np.exp(x)
y_log = np.log(np.abs(x) + 1)

plt.plot(x, y_square, label='y = x^2', color='b')
plt.plot(x, y_sin, label='y = sin(x)', color='r')
plt.plot(x, y_exp, label='y = e^x', color='g')
plt.plot(x, y_log, label='y = log(|x| + 1)', color='m')
plt.title('Plots of Multiple Functions')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.legend()
plt.show()
```
